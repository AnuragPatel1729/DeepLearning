```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import math

def is_prime(n):
    if n % 2 == 0 and n > 2:
        return False
    return all(n % i for i in range(3, int(math.sqrt(n)) + 1, 2))
```

```python
a = np.arange(1, 10**3)
foo = np.vectorize(is_prime)
pbools = foo(a)
primes = np.extract(pbools, a)
primes
```

```python
def get_coordinate(num):
    return num * np.cos(num), num * np.sin(num)
```

```python
get_coordinate(1)
```

```python
x = np.arange(1, 10)
get_coordinate(x)
```

```python
def create_plot(nums, figsize=8, s=None, show_annot=False):
    # nums = np.array(list(nums))
    x, y = get_coordinate(nums)
    plt.figure(figsize=(figsize, figsize))
    plt.axis("off")
    plt.scatter(x, y, s=s,)
    plt.show()
```

```python
create_plot(primes)
```
