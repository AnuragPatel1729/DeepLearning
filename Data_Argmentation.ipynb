```python
import tensorflow as tf
from tensorflow import keras
from keras.layers import Conv2D, Dense, Input, Flatten, MaxPooling2D, BatchNormalization, Dropout
from keras import Sequential
```

```python
from keras.preprocessing import image
```

```python
image_cat = image.load_img('/content/cat_1.jpg', target_size=(200, 200))
```

```python
image_cat
```

```python
data_gen = tf.keras.preprocessing.image.ImageDataGenerator(
    rotation_range=30,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
)
```

```python
type(image_cat)
```

```python
img = image.img_to_array(image_cat)
```

```python
img.shape
```

```python
input_batch = img.reshape(1,200,200,3)
```

```python
input_batch.shape
```

```python
i=0
for output in data_gen.flow(input_batch, batch_size=1,save_to_dir='/content/aug'):
  i = i+1
  if i == 10:
    break
```
